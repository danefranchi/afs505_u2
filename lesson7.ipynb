```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df.head(2)
```

```python
type(df['sepal_length'].values)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(df['petal_length'], df['petal_width'])
ax.axes.set_xlabel('petal Length')
ax.axes.set_ylabel('petal Width')
ax.set_title('petal length vs width')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
df2 = df.replace({'setosa': 'r', 'versicolor':'b', 'virginica':'g'})

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter('petal_length', 'petal_width',  c = 'species',  data = df2)

ax.axes.set_xlabel('petal Length')
ax.axes.set_ylabel('petal Width')
ax.set_title('petal length vs width')

plt.show()
```

```python
dfr =df2[ df2['species'] ==  'r']
dfb =df2[ df2['species'] ==  'b']
dfg =df2[ df2['species'] ==  'g']

species = ['setosa', 'versicolor', 'virginica']
height = [dfr['sepal_length'].mean(),dfb['sepal_length'].mean(),
          dfg['sepal_length'].mean()]

fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 3, 1)
ax.bar(species, height)


ax.axes.set_ylabel('sepal height')
ax.set_title('sepal height by species')
```

```python
fig = plt.figure()
axes = fig.subplots(2, 2)
plt.tight_layout()

axes[0,0].scatter('sepal_length', 'petal_length',  c = 'species',  data = df2)
axes[0,1].scatter('sepal_length', 'petal_width',  c = 'species',  data = df2)
axes[1,0].scatter('sepal_width', 'petal_length',  c = 'species',  data = df2)
axes[1,1].scatter('sepal_width', 'petal_width',  c = 'species',  data = df2)

axes[0,0].set_xlabel('sepal length')
axes[0,0].set_ylabel('petal length')
axes[0,0].set_title('sepal length x petal length')

axes[0,1].set_xlabel('sepal length')
axes[0,1].set_ylabel('petal width')
axes[0,1].set_title('sepal length x petal width')

axes[1,0].set_xlabel('sepal width')
axes[1,0].set_ylabel('petal length')
axes[1,0].set_title('sepal width x petal length')

axes[1,1].set_xlabel('sepal width')
axes[1,1].set_ylabel('petal width')
axes[1,1].set_title('sepal width x petal width')


plt.show()
```
