```python
import random

region = {'Southeast', 'Northeast', 'Midwest', 'Southwest', 'Northwest'}

```

```python
iris_df['region'] = np.random.choice(list(region), len(iris_df))
iris_df
```

```python
group = iris_df.groupby(['region'])
group.mean()
```

```python
iris_df['dev_stage'] = np.random.choice(['early', 'late'], iris_df.shape[0])
iris_df.head()
```

```python
group = iris_df.groupby(['region', 'dev_stage'])
group.mean()
```

```python
groups = iris_df.groupby(['species','region', 'dev_stage'])
groups.mean()
groups.count()
```
