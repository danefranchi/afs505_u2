```python
import numpy as np
```

```python
file = np.genfromtxt("iris.csv", dtype = 'float', delimiter = ',', skip_header = 1, usecols =(0,1,2,3))
print(file.shape)
#print(file)
```

```python
print(file[0:5])
```

```python
print(file[0,0])
print(file[0][0])
```

```python
#retrive first column
print(file[:,0])
```

```python
#retrieve last 2 cells of last row and last 2 cells of second to last row
print(file[148:,2:])
print(file[-2:,-2:])
```

```python
#retrieve all values greater than 4
print(file[file>4])
```

```python
#retrieve all rows that have at least one value of five
five = file > 5
rowsum = np.sum(five, axis = 1)

print(file[rowsum > 0,:].shape)
print(file[rowsum > 0,:])
```

```python
#calculate mean median stddev of first column
print(np.mean(file[:,0]))
```

```python
#calculate correlation coefficcient between 1st and 4th colum
#help(corrcoef)
print(np.corrcoef(file[:,0], file[:,3]))
```

```python
#resize add 5 rows
np.resize(file, (155,4)).shape
```

```python
#append add five new rows of random numbers between 30 and 40
rdm = np.random.randint(30, 40, (5, 4))
print(rdm)
x = np.append(file, rdm, axis = 0)
print(x)
```

```python
#fill last 5 rows with random number between 10 and 20
for i in range(150, 155):
    x[i,] = np.random.rand(4)*10 +10
print(x[-5:,])
```
