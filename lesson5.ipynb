```python
data3b = pd.read_csv('https://hivdb.stanford.edu/download/GenoPhenoDatasets/PI_DataSet.txt', sep = '\t')
data3b.head()
```

```python
#What are the variables?
#drug, fold change, postion of amino acids

#What are the observations?

#What are the values? fold resistance,
```

```python
data_mel = pd.melt(data3b, id_vars=['SeqID', 'FPV', 'ATV', 'IDV',
                                    'LPV', 'NFV','SQV','TPV', 'DRV'], 
                   var_name='Fold_change', value_name='location')
data_mel
```

```python
data_mel = data_mel[['SeqID', 'Fold_change', 'location', 'FPV', 'ATV', 'IDV',
                                    'LPV', 'NFV','SQV','TPV', 'DRV']]
data_mel
```

```python
data_mel2 = pd.melt(data_mel, id_vars=['SeqID', 'Fold_change', 'location'], 
                   var_name='Drug', value_name='amount')
data_mel2
```

```python
data_mel2.index = data_mel2["SeqID"]
data_mel2.head()
```
